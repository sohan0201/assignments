```python
#Question 1
ds=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
ds.head(10)
```

```python
sb.kdeplot(ds["MaxTemp"])
sb.kdeplot(ds["MinTemp"])
```

```python
sb.swarmplot(x=ds["Location"],y=ds["MinTemp"])
```

```python
sb.distplot(ds["Temp9am"])
```
