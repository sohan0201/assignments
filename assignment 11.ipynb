```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv")
df.head(10)
```

```python
dataset=df.isnull().sum()
print(dataset)
```

```python
df=df.drop(['id','date'],axis=1)
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df,train_size=0.6,test_size=0.4)
train_x=train.loc[:,train.columns !='price']
test_x=test.loc[:,test.columns !='price']
train_y=train['price']
test_y=test['price']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(train_x, train_y)
```

```python
test_pred=model.predict(test_x)
result=pd.DataFrame({'Actual_price':test_y,'predicted_price':test_pred})
```

```python
print(result)
```

```python
from sklearn.metrics import r2_score
print(r2_score(test_y,test_pred)*100)
```
